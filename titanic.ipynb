Build a function called `transform_column` that takes arguments for a `data_dict`, `column_name`, and `datatype`, and use it to transofm the columns that need transformation.

**To help identify potential sources of errors, explore the `set` of values in each column.**

````python
def transform_column(data_dict, column_name, datatype):
    converted_entries = []    
    for element in data_dict[column_name]:
        try:
            converted_entries.append(datatype(element))
        except:
            converted_entries.append(np.NaN)
    
    data_dict[column_name] = converted_entries
    
    return data_dict_columns
    
transform_column(data_dict_columns, 'PassengerId', int)
transform_column(data_dict_columns, 'Survived', int)
transform_column(data_dict_columns, 'Pclass', int)
transform_column(data_dict_columns, 'Age', int)
transform_column(data_dict_columns, 'SibSp', int)
transform_column(data_dict_columns, 'Parch', int)
transform_column(data_dict_columns, 'Ticket', int)
transform_column(data_dict_columns, 'Fare', float)
````

### B) for each of these values, what's the 90% confidence interval of the population mean?

* Create a function to find the confidence interval, and use it on each of the numeric columns.
* What's your interpretation of the interval?

````python
from scipy import stats

def confidence_interval90(observed_values_array, sample_size=1000):
    random_choice_mean = []
    for i in range(sample_size):
        mean = np.mean(np.random.choice(observed_values_array.dropna(), size=len(observed_values_array), replace=True))
        random_choice_mean.append(mean)
        CI90 = np.percentile(random_choice_mean, [5,95])
    return print("The confidence interval (90%) is {} and the actual observed mean is {}".format(CI90, observed_values_array.mean()))   

confidence_interval90(titanic_df['Age'])
confidence_interval90(titanic_df['Fare'])
confidence_interval90(titanic_df['Parch'])
confidence_interval90(titanic_df['SibSp'])
confidence_interval90(titanic_df['Survived'])
````
